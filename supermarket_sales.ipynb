```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('supermarket_sales.csv', encoding= 'unicode_escape')
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Occupation'], axis=1, inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
df.rename(columns= {'Unit price':'Unit_price'})
```

```python
df.describe()
```

```python
ax = sns.countplot(x = 'Gender',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Total'].sum().sort_values(by='Total', ascending=False)

sns.barplot(x = 'Gender',y= 'Total' ,data = sales_gen)
```

```python
ax = sns.countplot(data = df, x = 'Customer type', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
    
```

```python
sales_age = df.groupby(['Customer type'], as_index=False)['Total'].sum().sort_values(by='Total', ascending=False)
sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(x = 'Customer type',y= 'Total' ,data = sales_age)
```

```python
sales_state = df.groupby(['City'], as_index=False)['Quantity'].sum().sort_values(by='Quantity', ascending=False).head(10)

sns.set(rc={'figure.figsize':(5,6)})
sns.barplot(data = sales_state, x = 'City',y= 'Quantity')
```

```python
ax = sns.countplot(x = 'Product line',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product line'], as_index=False)['Rating'].sum().sort_values(by='Rating', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,6)})
sns.barplot(data = sales_state, x = 'Product line',y= 'Rating')
```

```python
ax = sns.countplot(data = df, x = 'Payment', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Invoice ID')['Quantity'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
